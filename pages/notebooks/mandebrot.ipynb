```python
from functools import partial

from mandelbrot import (
    create_fractal,
    create_fractal_gpu,
    create_fractal_jit,
    show_fractal,
)
```

```python
resolution = (1440, 3440)
iter_small = 10
iter_medium = 10_000
iter_large = 1_000_000
view_rectangle = (-2.5, 1.5, -1.2, 1.2)
```

```python
def show_size(iters: int):
    """Create a function to show the fractal with a specific number of iterations."""
    return partial(
        show_fractal,
        max_iters=iters,
        resolution=resolution,
        view=view_rectangle,
    )


show_quick = show_size(iter_small)
show_medium = show_size(iter_medium)
show_long = show_size(iter_large)
```

```python
for worker in [
    create_fractal,
    create_fractal_jit,
    create_fractal_gpu,
]:
    show_quick(worker)
```

```python
for worker in [
    create_fractal_jit,
    create_fractal_gpu,
]:
    show_medium(worker)
```

```python
show_fractal(
    fractal_worker=create_fractal_gpu,
    max_iters=10_000_000,
    resolution=(14400, 34400),
    view=view_rectangle,
)
```
