```python
import mplfinance as mpf
import yfinance as yf
from rich import print as rprint
```

```python
ticker_name = "CDR.WA"
yf_ticker = yf.Ticker(ticker_name)
rprint(yf_ticker.info["longBusinessSummary"])
```

```python
period = "6mo"
interval = "1d"
historical_data = yf_ticker.history(
    period=period,
    interval=interval,
)
historical_data.tail(10)
```

```python
mpf.plot(
    historical_data,
    type="candle",
    mav=(10, 24),
    volume=True,
    style="yahoo",
    figratio=(3, 2),
    figscale=1.5,
    show_nontrading=False,
    title=yf_ticker.info["longName"],
    tight_layout=True,
    warn_too_much_data=10_000,
)
```
