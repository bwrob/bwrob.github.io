```python
# | echo: false
# | output: asis
# | collapse: true

import mplfinance as mpf
import yfinance as yf
from IPython.display import Markdown, display

tabs = ["CDR.WA", "PKO.WA"]
display(Markdown(":::{.panel-tabset}"))

for tab in tabs:
    display(Markdown("\n"))
    display(Markdown(f"## {tab}"))
    display(Markdown("\n"))
    display(Markdown(f"Hello {tab}"))
    period = "6mo"
    interval = "1d"
    yf_ticker = yf.Ticker(tab)
    historical_data = yf_ticker.history(
        period=period,
        interval=interval,
    )
    mpf.plot(
        historical_data,
        type="candle",
        mav=(10, 24),
        volume=True,
        style="yahoo",
        figratio=(3, 2),
        figscale=1,
        show_nontrading=False,
        title=yf_ticker.info["longName"],
        tight_layout=True,
        warn_too_much_data=10_000,
    )

display(Markdown(":::"))
```
